## Pie Charts

```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40]

plt.pie(x)

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40]
y = ['C', 'C++', 'Python', 'Java']
ex = [0.4, 0.0, 0.0, 0.0]

plt.pie(x, labels = y, explode = ex)

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40]
y = ['C', 'C++', 'Python', 'Java']
color = ["cyan", 'grey', 'red', 'pink']
ex = [0.2, 0.0, 0.0, 0.0]

plt.pie(x, labels = y, explode = ex, colors = color, autopct = "%0.2f%%") # autopct is used to display percentages

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40]
y = ['C', 'C++', 'Python', 'Java']
color = ["cyan", 'grey', 'red', 'pink']
ex = [0.2, 0.0, 0.0, 0.0]

plt.pie(x, labels = y, explode = ex, colors = color, autopct = "%0.2f%%", shadow = True, radius = 1.5)

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40]
y = ['C', 'C++', 'Python', 'Java']
color = ["cyan", 'grey', 'red', 'pink']
ex = [0.2, 0.0, 0.0, 0.0]

plt.pie(x, labels = y, explode = ex, colors = color, autopct = "%0.2f%%", shadow = True, radius = 1, labeldistance = 1.2, startangle = 90,
       textprops = {
           "fontsize" : 10
       }, 
        counterclock = False, 
        wedgeprops = {
           "linewidth" : 5, 
            "width" : 1,
            "edgecolor" : 'm'
        },
        center = (3, 6)
    )

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40]
y = ['C', 'C++', 'Python', 'Java']
color = ["cyan", 'grey', 'red', 'pink']
ex = [0.2, 0.0, 0.0, 0.0]

plt.pie(x, labels = y, explode = ex, colors = color, autopct = "%0.2f%%", shadow = True, radius = 1, labeldistance = 1.2, startangle = 90,
       textprops = {
           "fontsize" : 10
       }, 
        counterclock = False, 
        wedgeprops = {
           "linewidth" : 5, 
        },
        rotatelabels = False
    )


plt.title("Pie Charts")
plt.legend(loc = 1) # loc : location
plt.show()
```

```python
import matplotlib.pyplot as plt

x1 = [10, 20, 30, 40]
x2 = [40, 30, 20, 10]
y = ['C', 'C++', 'Python', 'Java']
c = ['r', 'b', 'g', 'y']

plt.pie(x1, labels = y, radius = 1.5)
plt.pie(x2, radius = 0.5, colors = c)

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40]
y = ['C', 'C++', 'Python', 'Java']
c = ['r', 'b', 'g', 'y']

plt.pie(x, labels = y, radius = 1.5)
cr = plt.Circle(xy = (0, 0), radius = 1, facecolor = 'w')

plt.gca().add_artist(cr)
plt.show()
```
