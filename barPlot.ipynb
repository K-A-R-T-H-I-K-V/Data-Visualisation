## Bar Plot

```python
import matplotlib.pyplot as plt

x = ['Python', 'C', 'C++', 'Java']
y = [85, 70, 60, 82]

plt.bar(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = ['Python', 'C', 'C++', 'Java']
y = [85, 70, 60, 82]
colors = ['cyan', 'black', 'indigo', 'purple']

plt.xlabel('Languages', fontsize = 10)
plt.ylabel('Popularity ratio', fontsize = 10)
plt.title('Anamica')

plt.bar(x, y, width = 0.3, color = colors, align = "edge") # align takes 'edge' or 'centre'
plt.show()
```

```python
import matplotlib.pyplot as plt

x = ['Python', 'C', 'C++', 'Java']
y = [85, 70, 60, 82]

plt.xlabel('Languages', fontsize = 10)
plt.ylabel('Popularity ratio', fontsize = 10)
plt.title('Anamica')

plt.bar(x, y, width = 0.3, color = 'cyan', edgecolor = 'black', linewidth = 1.5, linestyle = ':', alpha = 0.5) # alpha ranges from 0 to 1, reduces the brightness
plt.show()
```

```python
import matplotlib.pyplot as plt

x = ['Python', 'C', 'C++', 'Java']
y = [85, 70, 60, 82]
z = [25, 50, 72, 40]

plt.xlabel('Languages', fontsize = 10)
plt.ylabel('Popularity ratio', fontsize = 10)
plt.title('Anamica')

plt.bar(x, y, width = 0.3, color = 'cyan', edgecolor = 'black', label = "Popularity1") 
plt.bar(x, z, width = 0.3, color = 'purple', edgecolor = 'black', label = "Popularity2") 

plt.legend() # For label
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = ['Python', 'C', 'C++', 'Java']
y = [85, 70, 60, 82]
z = [25, 50, 72, 40]

# Avoid overlap
width = 0.2
p = np.arange(len(x))
p1 = [j + width for j in p]

plt.xlabel('Languages', fontsize = 10)
plt.ylabel('Popularity ratio', fontsize = 10)
plt.title('Anamica')

plt.bar(p, y, width, color = 'green', edgecolor = 'black', label = "Popularity1") 
plt.bar(p1, z, width, color = 'indigo', edgecolor = 'black', label = "Popularity2") 

plt.legend() # For label
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = ['Python', 'C', 'C++', 'Java']
y = [85, 70, 60, 82]
z = [25, 50, 72, 40]

# Avoid overlap
width = 0.2
p = np.arange(len(x))
p1 = [j + width for j in p]

plt.xlabel('Languages', fontsize = 10)
plt.ylabel('Popularity ratio', fontsize = 10)
plt.title('Anamica')

plt.bar(p, y, width, color = 'red', edgecolor = 'black', label = "Popularity1") 
plt.bar(p1, z, width, color = 'purple', edgecolor = 'black', label = "Popularity2") 

plt.xticks(p + width / 2, x, rotation = 1)
plt.legend() # For label
plt.show()
```

```python
import matplotlib.pyplot as plt

x = ['Python', 'C', 'C++', 'Java']
y = [85, 70, 60, 82]
z = [25, 50, 72, 40]

plt.xlabel('Languages', fontsize = 10)
plt.ylabel('Popularity ratio', fontsize = 10)
plt.title('Anamica')

plt.barh(x, y, color = 'cyan', edgecolor = 'black', label = "Popularity1") 
plt.bar(x, z, color = 'purple', edgecolor = 'black', label = "Popularity2") 

plt.legend() # For label
plt.show()
```
