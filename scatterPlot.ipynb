## Scatter Plot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6, 7]
y = [2, 3, 1, 4, 5, 3, 6]

plt.scatter(x, y)

plt.show()
```

```python
import matplotlib.pyplot as plt

day = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday',' Saturday']
num = [2, 3, 1, 4, 5, 3, 6]
colors = ['violet', 'indigo', 'b', 'g', 'y', 'orange', 'r']
sizes = [600, 25, 490, 60, 230, 150, 200]

plt.scatter(day, num, c = colors, s = sizes, alpha = 0.5) # c - color, s - size, alpha reduces transparency

plt.xlabel("Days")
plt.ylabel("Frequency")
plt.title("Scatter Plot", fontsize = 15)

plt.show()
```

```python
import matplotlib.pyplot as plt

day = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday',' Saturday']
num = [2, 3, 1, 4, 5, 3, 6]
colors = ['violet', 'indigo', 'b', 'g', 'y', 'orange', 'r']
sizes = [600, 250, 490, 600, 230, 150, 200]

plt.scatter(day, num, c = colors, s = sizes, marker = '*', edgecolor = 'black', linewidth = 1) # c - color, s - size, alpha reduces transparency

plt.xlabel("Days")
plt.ylabel("Frequency")
plt.title("Scatter Plot", fontsize = 15)

plt.show()
```

```python
import matplotlib.pyplot as plt

day = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday',' Saturday']
num = [2, 3, 1, 4, 5, 3, 6]
colors = [10, 30, 49, 29, 80, 75, 45]
sizes = [600, 25, 490, 60, 230, 150, 200]

plt.scatter(day, num, c = colors, s = sizes, cmap = "viridis") # cmap - color mapping
plt.colorbar()

plt.xlabel("Days", fontsize = 15)
plt.ylabel("Frequency", fontsize = 15)
plt.title("Scatter Plot", fontsize = 15)

plt.show()
```

```python
import matplotlib.pyplot as plt

day = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday',' Saturday']
num1 = [2, 3, 1, 4, 5, 3, 6]
num2 = [20, 30, 10, 40, 50, 30, 60]
colors = [10, 30, 49, 29, 80, 75, 45]
sizes = [600, 25, 490, 60, 230, 150, 200]

plt.scatter(day, num1, c = colors, s = sizes, cmap = "Oranges") 
t1 = plt.colorbar()
t1.set_label("Color Bar - Oranges")

plt.scatter(day, num2, c = colors, s = sizes, cmap = "Greens") 
t2 = plt.colorbar()
t2.set_label("Color Bar - Greens", fontsize = 10)

plt.xlabel("Days", fontsize = 10)
plt.ylabel("Frequency", fontsize = 10)
plt.title("Scatter Plot", fontsize = 10)

plt.show()
```
